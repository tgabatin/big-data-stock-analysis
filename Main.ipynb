```python
# Give me a count for the files we have so we can loop through them, and some fries on the side

import pathlib

stock_count, etf_count = 0, 0
for path in pathlib.Path("data/stocks").iterdir():
    if path.is_file():
        stock_count += 1

print("Stock file count: ", stock_count)

for path in pathlib.Path("data/etfs").iterdir():
    if path.is_file():
        etf_count += 1

print("Etf file count: ", etf_count)
```

```python
# Load the files into one big list of dataframes and normalize dates according to pandas
# yahoo finance package dealt with pandas date normalization  

stock_df_list = []
etf_df_list = []

for i in range(len(nomenclature)):
    if nomenclature['ETF'][i] == 'Y':
        etf_df_list.append(pd.read_csv('data/etfs/' + nomenclature['Symbol'][i] +'.csv'))
    else:
        stock_df_list.append(pd.read_csv('data/stocks/' + nomenclature['Symbol'][i] +'.csv'))
```

```python
# Seperate stocks into 3 types and drop stocks with history < 3 years, 2018 and younger
# Type I: Longer than 25 years
# Type 2: 25y and younger
# Type 3: After 2009 

stocks_relevant = [s for s in stock_df_list if s['Date'][0] < '2018-01-01']
stocks_type_3 = [s for s in stocks_relevant if s['Date'][0] > '2009-12-31']
stocks_type_2 = [s for s in stocks_relevant if s['Date'][0] > '1994-12-31']
stocks_type_1 = [s for s in stocks_relevant if s['Date'][0] < '1994-12-31']
```
