```python
# Import statements
import pandas as pd
#import pyspark
#from pyspark import SparkContext
```

```python
# Load symbols_valid_meta to get reference Stocks and ETFs with full name. 

nomenclature = pd.read_csv('data/symbols_valid_meta.csv')
nomenclature
```

```python
# We need to drop special characters like '-' and '$' from the dataframe
# When we read the csv into panda df, '-' for example turned into a '-', 
# which now is a problem when reading the files in
```

```python
# Give me a count for the files we have so we can loop through them, and some fries on the side

import pathlib

stock_count, etf_count = 0, 0
for path in pathlib.Path("data/stocks").iterdir():
    if path.is_file():
        stock_count += 1

print("Stock file count: ", stock_count)

for path in pathlib.Path("data/etfs").iterdir():
    if path.is_file():
        etf_count += 1

print("Etf file count: ", etf_count)
```

```python
# Load the files into big df and normalize dates according to pandas

stock_df_list = []
etf_df_list = []

for i in range(len(nomenclature)):
    if nomenclature['ETF'][i] == 'Y':
        etf_df_list.append(pd.read_csv('data/etfs/' + nomenclature['Symbol'][i] +'.csv'))
    else:
        stock_df_list.append(pd.read_csv('data/stocks/' + nomenclature['Symbol'][i] +'.csv'))
```
